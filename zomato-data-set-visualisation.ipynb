```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

# Reading CSV

```python
df = pd.read_csv('/kaggle/input/zomato-dataset/zomato.csv')
df.head()     #is used to represent first five rows
```

```python
df.shape   #to get detail that how many rows and columns are there
```

```python
df.columns
```

```python
df=df.drop(['url', 'address', 'phone', 'menu_item' ,'dish_liked', 'reviews_list'],axis=1)
df.head()
```

```python
df.info()    #infomation about dataset (datatype,if there are null values or not)
```

### Dropping Duplicates

```python
df.drop_duplicates(inplace = True)     #it will remove the rows which are repeated or more than 1 times present data
df.shape
```

## Cleaning Rate Column

```python
df['rate'].unique()
```

## Removing "NEW" , "-" and "/5" from Rate column

```python
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')       #it means if it is string then it will be divided into two parts with / and the part at 0 index will be return as float datatype
        value=value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

## Filling Null Values in Rate Column with Mean

```python
df.rate.isnull().sum()    #isnull()is used to know whether there are null values or not.
```

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)      #fillna keyword is used to fill the place of null spaces,here mean is used to fill
df['rate'].isnull().sum()
```

```python
df.info()
```

### Dropping Null values

```python
df.dropna(inplace = True)     #dropna is used to drop the null values in the dataset.
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)': 'Cost2plates', 'listed_in(type)': 'Type'},inplace=True)
#above  is used to rename the long words with a shorter one
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
#in these two 'location' and 'listed in city' containing cities name in common so we can drop one which is not necessary
df.drop(['listed_in(city)'],axis=1)
```

```python
df['Cost2plates'].unique()
```

```python
#most of the values are containing comma which make it string not integer, to remove these commas we will create a function
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

###  Cleaning Rest Type Column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
#those restaurants which are having less than 1000 no. of quantities will be stored in the above variable
rest_types_lessthan1000
```

### Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_types(value):
    if (value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(handle_rest_types)
df['rest_type'].value_counts()
```

```python
df.head()
```

###  Cleaning Location Column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)

location_lessthan500 = location[location<500]

def handle_location(value):
    if(value in location_lessthan500):
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

### Cleaning Cuisines Column 

```python
cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if (value in cuisines_lessthan100):
        return 'others'
    else:
        return value
df['cuisines']= df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```
